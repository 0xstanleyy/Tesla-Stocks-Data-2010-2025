```python
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv(r"C:\Users\IZOSE  OMELIME\Downloads\Tesla Stock Data 2025\TSLA_2010-06-29_2025-02-13.csv")
df['Date'] = pd.to_datetime(df['Date'])
```

```python
lp = sns.lineplot(x= 'Date', y='Close', data = df)
lp.set_title('Tesla Stocks Closing Price Over Time (2010-2025)', fontsize = 14)
lp.set_ylabel('')
lp.set_xlabel('')
lp.set_yticklabels([])
lp.set_xlim(left=None, right=pd.to_datetime('2025-12-31'))
plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(
    nrows=2, 
    ncols=1, 
    sharex=True,          
    figsize=(10, 7), 
    gridspec_kw={'height_ratios': [3, 1]} # Gives the price chart 3x more vertical space
)


ax1.plot(df.index, df['Close'], color='blue', label='Closing Price')
ax1.set_ylabel('Closing Price (USD)', fontsize=12)
ax1.set_title('Tesla Stocks: Closing Price vs. Trading Volume Over Time', fontsize=14)

ax1.grid(True, linestyle='--', alpha=0.6)

ax2.bar(df.index, df['Volume'], color='gray', alpha=0.7, label='Trading Volume')
ax2.set_ylabel('Volume', fontsize=12)
ax2.xaxis.set_visible(False)
ax2.grid(True, linestyle='--', alpha=0.6)

plt.tight_layout() 
plt.show()
```
